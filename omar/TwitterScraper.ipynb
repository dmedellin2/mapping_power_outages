```python
import json
import pandas as pd
import re
import datetime
from twitterscraper import query_tweets

df = pd.read_json('/Users/osmiley/Documents/GA/twitterscraper/power_outage_tweets_la.json')
df
```

```python
cities_list =['columbus', 'dallas', 'houston', 'LA',]
```

```python
for city in cities_list:
    df_tweets = pd.DataFrame(columns=['id','text','timestamp','user','location'])
    tweet_list = query_tweets(f'"outage" OR "power\'s out" OR "power out" OR "power outage" OR "pwr out" OR "power failure" OR "without power" OR "power is out" OR "electric is out" OR "electricity out" OR "electricity\'s out" OR "without electricity" OR "lost electricity" OR "lost power" OR "#poweroutage" -filter:retweets near:"{city}" within:10mi',
                              begindate = datetime.date(2014,1,1),
                              enddate = datetime.date(2019,4,15),
                              poolsize = 10)
    for row, tweet in enumerate(tweet_list):
        df_tweets.loc[row,'id'] = tweet[id]
        df_tweets.loc[row,'text'] = tweet.text
        df_tweets.loc[row,'timestamp'] = tweet.timestamp
        df_tweets.loc[row,'user'] = tweet.user
        df_tweets.loc[row,'location'] = city
    df_tweets.to_csv(f'../datasets/tweets/{city}_tweets.csv')
```
